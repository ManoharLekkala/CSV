```python
import csv 
from collections import Counter
titles = list(csv.DictReader(open('titles.csv')))
import matplotlib.pyplot as plt

by_year = Counter(t['year'] for t in titles)
print(by_year.most_common(4))

```

```python
data = (by_year.most_common(4))
data = sorted(data)
x = [c[0] for c in data]
y = [c[1] for c in data]
plt.plot(x,y)
```

```python
# DEFAULT DICTIONARY
from collections import defaultdict
dd = defaultdict(list)
for row in titles:
    dd[row['year']].append(row['title'])
    
xx = []
yy = []
for k , v in dd.items():
    xx.append(k)
    yy.append(len(v))
    
plt.plot(sorted(xx) , yy)
```
